Using the example from the book, we can plot inline like so: 

```python
import pandas as pd
names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,83,77,78,95]
GradeList = zip(names,grades)
df = pd.DataFrame(data = GradeList,
 columns=['Names', 'Grades'])
%matplotlib inline
df.plot()
```

However, to plot Bob's score in a way that makes sense (to emphasize hsi high score), we can add annotations on a bar plot as well:

```python
df
df_bob = df[df['Names'] == 'Bob']



plt = df.plot.bar(x = "Names", y = 'Grades')



displayText = "my annotation"
xloc = 0.05
yloc =  df_bob['Grades'].item()
xtext = 8
ytext = 150
plt.annotate("Wow!",
 xy=(xloc, yloc),
 arrowprops=dict(facecolor='black',
 shrink=0.05),
 xytext=(xtext,ytext),
 xycoords=('axes fraction', 'data'),
 textcoords='offset points')
```
