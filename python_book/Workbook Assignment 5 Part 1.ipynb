For this week's assigment, we will explore pandas and Python's random functions for us to sample a dataframe

```python
import pandas as pd
import numpy as np
Location = "gradedata.csv"
df = pd.read_csv(Location)
df.tail()
```

This is the code to randomly sample a dataframe. Although it's simple to change the amount of rows to resample, it's important to understand what the function is doing underneath. We are leveraging the Python random functions to shuffle in the indices of our dataframe and return a dataframe (take) basked on the shuffle indices. 

```python
df.take(np.random.permutation(len(df))[:500])
```

We can check whether the function is wokring by having the function loop a few times, we see that the observations shift for each print statement

```python
for iterations in range(6): 
    print(df.take(np.random.permutation(len(df))[:500]))
```
